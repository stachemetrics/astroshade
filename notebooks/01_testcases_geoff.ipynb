```python
# suggestions from another AI session
from google import genai
from pydantic import BaseModel, Field
from enum import Enum

client = genai.Client()

# 1. Define the strictly allowed Goldwell products
class LightenerEnum(str, Enum):
    SILKLIFT_STRONG = "SilkLift Strong"
    SILKLIFT_ZERO_AMMONIA = "SilkLift Zero Ammonia"
    OXYCUR_PLATIN = "Oxycur Platin"
    NONE = "None"

class DeveloperEnum(str, Enum):
    VOL_10 = "3% (10 Vol)"
    VOL_20 = "6% (20 Vol)"
    VOL_30 = "9% (30 Vol)"
    VOL_40 = "12% (40 Vol)"
    COLORANCE_LOTION = "Colorance Lotion 2%"

# 2. Define the exact JSON structure Matt expects to see
class HairFormulation(BaseModel):
    starting_level_assessment: int = Field(description="The assumed starting level of the hair (1-10)")
    lightener_choice: LightenerEnum
    lightener_developer: DeveloperEnum
    lightener_ratio: str = Field(description="e.g., 1:1 or 1:2")
    toner_formula: str = Field(description="The exact Colorance or Topchic shades used, e.g., 10V + 10P")
    toner_developer: DeveloperEnum
    processing_time: str = Field(description="Recommended processing time in minutes")
    application_technique: str = Field(description="Step-by-step application instructions")
    color_theory_reasoning: str = Field(description="Crucial: Explain WHY this formulation works for this specific case")

# 3. Call the model using Structured Outputs
print("Generating formulation...")
response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents="Client wants to be super bright blonde all over, like an icy Scandi blonde. I hate seeing any yellow or gold. Starting point is a virgin Level 7.",
    config={
        'response_mime_type': 'application/json',
        'response_schema': HairFormulation,
        'temperature': 0.1 # Keep this low so it relies on facts, not creativity
    },
)

print(response.text)
```
